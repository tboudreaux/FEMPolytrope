```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use(pubStyle)
```

```python
exactSolution = lambda xi: np.sin(xi)/xi
```

```python
initialGuess = pd.read_csv("../laneEmdenNonlinear_presolve.csv")
mfemSolution = pd.read_csv("../laneEmdenNonlinear.csv")
```

```python
XI = np.linspace(0.001, 4, 1000)
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
ax.plot(initialGuess.xi.values, initialGuess.u.values, linestyle='dashed', color='blue', label='Initial Guess')
ax.plot(XI, exactSolution(XI), color='green', label='Exact Soluition')
ax.axhline(y=0, color='grey', alpha=0.25, linestyle='dashed')
ax.axvline(x=np.pi, color='grey', alpha=0.25, linestyle='dashed')
ax.plot(mfemSolution.xi.values, mfemSolution.u.values, color='black', label='MFEM Solution')
ax.set_xlabel(r'$\xi$')
ax.set_ylabel(r'$\theta$')
ax.legend(frameon=True)
```
